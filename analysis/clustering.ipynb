# Clustering - Marketing Analytics

Pipeline de clusterizacao usando Autoencoder + K-Means.

```python
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
```

```python
from marketing_analytics.cluster import calcular_elbow, treinar_kmeans
from marketing_analytics.features import padronizar_features
from marketing_analytics.loading import carregar_dados, extrair_features, limpar_dados
from marketing_analytics.reduce import (
    construir_autoencoder,
    extrair_encoder,
    treinar_autoencoder,
)
```

```python
df = carregar_dados('../../data/Marketing_data.csv')
df_limpo, medianas = limpar_dados(df)
features_brutas = extrair_features(df_limpo)
features_pad, scaler = padronizar_features(features_brutas)
```

```python
autoencoder = construir_autoencoder(input_dim=17)
treinar_autoencoder(features_pad, autoencoder, epochs=50, batch_size=128)
encoder = extrair_encoder(autoencoder)
features_comprimidas = encoder.predict(features_pad)
```

```python
inercias = calcular_elbow(features_comprimidas, k_max=10)
plt.plot(range(1, 11), inercias, marker='o')
plt.xlabel('Numero de clusters (k)')
plt.ylabel('Inercia')
plt.title('Metodo do Cotovelo')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
kmeans = treinar_kmeans(features_comprimidas, n_clusters=4)
labels = kmeans.labels_
print(f'Silhouette: {silhouette_score(features_comprimidas, labels):.4f}')
```
